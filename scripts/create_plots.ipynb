```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Scoring Power
# Data
data = pd.read_csv("../data/CASF-2016/power_scoring/results_power_scoring.csv")

# Remove everything that missed samples
data = data[data["Number of favorable sample (N):"] == 285]
data = data.sort_values(by=["Pearson correlation coefficient (R)"])

lower_bounds = []
upper_bounds = []

r_values = list(data["Pearson correlation coefficient (R)"])
conf_int = data["Confidence interval of 90% Pearson R"]
for value in conf_int:
    tmp = value.split(sep="~")
    lower_bounds.append(float(tmp[0][1:-1]))
    upper_bounds.append(float(tmp[1][1:-2]))

name = []
for i in range(len(r_values)):
    name.append(f"{list(data['Datatype'])[i]} {list(data['Modeltype'])[i]} {list(data['Scoretype'])[i]}")

box_width = 0.8  # Adjust this value to control the width of the boxes

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10,10))

# Plot the custom "box" with a line inside
for y, (lower, upper) in enumerate(zip(lower_bounds, upper_bounds)):
    rect = plt.Rectangle((lower, y + 0.4 - box_width / 2), upper - lower, box_width, fill=True, color='#0063A6', alpha=0.9)
    ax.add_patch(rect)
    ax.plot([lower + (upper - lower) / 2] * 2, [y + 0.4 - box_width / 2, y + 0.4 + box_width / 2], color='white')

# Set labels and title
ax.set_yticks(range(len(r_values)))
ax.set_yticklabels(name)

ax.set_xticks([-0.5 + i * 0.1 for i in range(16)])
ax.set_xlabel('Values')
ax.set_title('Scoring power of all complete samples')

# Show the plot
plt.tight_layout()
plt.show()
```
